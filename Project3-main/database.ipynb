```python
import sqlite3
import pandas as pd
```

```python
weekdaysdb = pd.read_csv('./Resources/weekdays_r.csv')
weekendsdb = pd.read_csv('./Resources/weekends_r.csv')
```

```python
weekdaysdb.head()
```

```python
df1 = weekdaysdb[weekdaysdb['loc']=="Paris"]
df2 = weekendsdb[weekendsdb['loc']=="Paris"]
df3= pd.concat([df1, df2], ignore_index=True)
```

```python
df3.head()
```

```python
df4 = df3.drop('loc', axis=1)
```

```python
df4['loc']='paris'
```

```python
df4.head()
```

```python
conn = sqlite3.connect('abdata.sqlite') 
c = conn.cursor()

c.execute('''
          CREATE TABLE IF NOT EXISTS paris
          ([_id] string PRIMARY KEY, [attr_index] float, [attr_index_norm] float, [bedrooms] integer,
           [biz] integer, [cleanliness_rating] integer, [dist] float, [guest_satisfaction_overall] integer,
           [host_is_superhost] boolean, [lat] float, [lng] float, [loc] string, [metro_dist] float,
            [multi] integer, [person_capacity] integer, [realSum] float, [rest_index] float, [rest_index_norm] float,
            [room_type] string)
          ''')
conn.commit()
```

```python
df4.to_sql('paris', conn, if_exists='append', index = False)
```

```python
conn = sqlite3.connect('abdata.sqlite') 
c = conn.cursor()

c.execute('''
          CREATE TABLE IF NOT EXISTS weekends
          ([_id] string PRIMARY KEY, [attr_index] float, [attr_index_norm] float, [bedrooms] integer,
           [biz] integer, [cleanliness_rating] integer, [dist] float, [guest_satisfaction_overall] integer,
           [host_is_superhost] boolean, [lat] float, [lng] float, [loc] string, [metro_dist] float,
            [multi] integer, [person_capacity] integer, [realSum] float, [rest_index] float, [rest_index_norm] float,
            [room_type] string)
          ''')
          
c.execute('''
          CREATE TABLE IF NOT EXISTS weekdays
          ([_id] string PRIMARY KEY, [attr_index] float, [attr_index_norm] float, [bedrooms] integer,
           [biz] integer, [cleanliness_rating] integer, [dist] float, [guest_satisfaction_overall] integer,
           [host_is_superhost] boolean, [lat] float, [lng] float, [loc] string, [metro_dist] float,
            [multi] integer, [person_capacity] integer, [realSum] float, [rest_index] float, [rest_index_norm] float,
            [room_type] string)
          ''')

c.execute('''
          CREATE TABLE IF NOT EXISTS combined
          ([_id] string PRIMARY KEY, [attr_index] float, [attr_index_norm] float, [bedrooms] integer,
           [biz] integer, [cleanliness_rating] integer, [dist] float, [guest_satisfaction_overall] integer,
           [host_is_superhost] boolean, [lat] float, [lng] float, [loc] string, [metro_dist] float,
            [multi] integer, [person_capacity] integer, [realSum] float, [rest_index] float, [rest_index_norm] float,
            [room_type] string)
          ''')
                     
conn.commit()
```

```python
# write the data to a sqlite table
weekdaysdb.to_sql('weekdays', conn, if_exists='append', index = False)
```

```python
weekendsdb.to_sql('weekends', conn, if_exists='append', index = False)
```

```python
weekdaysdb.to_sql('combined', conn, if_exists='append', index = False)
weekendsdb.to_sql('combined', conn, if_exists='append', index = False)
```
